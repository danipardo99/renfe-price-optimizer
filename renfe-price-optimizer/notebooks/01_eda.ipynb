Notebook de exploración. Aquí SOLO exploramos. La lógica estable vive en `src/renfe_optimizer/features.py`.

```python
import matplotlib.pyplot as plt
import pandas as pd

from renfe_optimizer.features import build_features

df_raw = pd.read_csv('../data/raw/renfe_sample.csv')
df_raw.shape, df_raw.head()
```

```python
df = build_features(df_raw)
print(f'Filas Madrid-Barcelona con precio: {len(df)}')
df.describe(include='all').T
```

## Curva precio-antelación

```python
curva = df.groupby('dias_anticipacion')['price'].mean().reset_index()
fig, ax = plt.subplots(figsize=(9,4))
ax.plot(curva['dias_anticipacion'], curva['price'])
ax.invert_xaxis()
ax.set_xlabel('Días de antelación')
ax.set_ylabel('Precio medio (€)')
ax.set_title('Curva precio-antelación — Madrid-Barcelona 2019')
plt.show()
```
